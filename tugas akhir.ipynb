```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Menghilangkan kolom 'tanggal' dan 'nama_produk' karena tidak numerik
df_numerik = df.drop(columns=['tanggal', 'nama_produk'])

# Melakukan hierarchical clustering
linked = linkage(df_numerik, 'ward')

# Membuat dendogram
plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=df['nama_produk'].values,
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Dendogram Penjualan Martabak')
plt.xlabel('Produk')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Menghilangkan kolom 'tanggal' dan 'nama_produk' karena tidak numerik
df_numerik = df.drop(columns=['tanggal', 'nama_produk'])

# Normalisasi data
scaler = StandardScaler()
df_numerik_scaled = scaler.fit_transform(df_numerik)

# Melakukan clustering dengan K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(df_numerik_scaled)

# Menambahkan kolom cluster ke DataFrame asli
df['cluster'] = clusters

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
plt.scatter(df['jumlah'], df['harga_satuan'], c=df['cluster'], cmap='viridis', label='Cluster')
plt.xlabel('Jumlah Terjual')
plt.ylabel('Harga Satuan')
plt.title('Cluster Plot Jumlah Terjual vs Harga Satuan')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```
